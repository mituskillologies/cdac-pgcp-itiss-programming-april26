Given Input:

tech_vedas = {"Aarav", "Priya", "Rahul", "Sneha"}
data_sutras = {"Rahul", "Sneha", "Vikram", "Anjali"}

Expected Output:

{'Anjali', 'Priya', 'Vikram', 'Aditi', 'Aarav'} 
# Note: The order of elements in the output set 
# may vary since sets are unordered.

```python
tech_vedas = {"Aarav", "Priya", "Rahul", "Sneha"}
data_sutras = {"Rahul", "Sneha", "Vikram", "Anjali"}

all_employees = tech_vedas.union(data_sutras) 

double_agents = tech_vedas.intersection(data_sutras) 

final_roster = all_employees.difference(double_agents) 

final_roster.add("Aditi")

print(final_roster)
```
