This is a list of some metallic elements.
metals = [ 'silver','gold', 'copper','aluminium','iron','gold', 'copper','iron', 'copper', 'gold','gold','copper','aluminium']
• Make a new list that is almost identical to the
metals list: the new contains the same items, in
the same order, except that it does NOT contain
the item 'copper'.

```python
metals = [ 'silver','gold', 'copper','aluminium',
          'iron','gold', 'copper','iron', 'copper', 
          'gold','gold','copper','aluminium']
```

```python
new_metals = []
for metal in metals: 
    if metal != 'copper':
        new_metals.append(metal)
```

```python
new_metals
```

```python
new_metals = [metal for metal in metals if metal != 'copper']
```

```python
new_metals
```
