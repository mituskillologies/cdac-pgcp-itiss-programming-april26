```python
import socket
```

```python
s = socket.socket()
s.connect(('localhost', 1235))

data = input("Enter the data:")
s.send(data.encode('utf-8'))
print("Data Received:", s.recv(20))
s.close()
```

```python
# create socket object
s = socket.socket()
# bind the socket to IP and port 
s.bind(('192.168.74.126', 1236))
# listen for 5 client connections
s.listen(5)
while True:
    print("Waiting for client connection...")
    c, addr = s.accept()  # create client object
    print("Go connection from", addr)
    data = c.recv(2000)
    data = data.decode('utf-8')
    c.send(data.swapcase().encode('utf-8'))
    c.close()
```
