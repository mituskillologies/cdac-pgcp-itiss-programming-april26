```python
a = [56,21,35,47,68,73,17]
```

```python
list(enumerate(a))
```

```python
dict(list(enumerate(a)))
```

```python
for i, v in list(enumerate(a)):
    print(i,v)
```

```python
d = dict((i,v) for i, v in list(enumerate(a)))
```

```python
car = {"brand": "Ford"}
# Key exists, returns current value
print(car.setdefault("brand", "Toyota")) # Output: Ford

# Key doesn't exist, inserts and returns new value
print(car.setdefault("color", "red"))

# Output: red
print(car)
```
