#### search

```python
import re
```

```python
text = "Call me at 555-1234."
```

```python
match = re.search(r"\d{3}-\d{4}", text)
```

```python
if match:
    print(f"Match found:{match.group()}")
```

#### sub

```python
text = "I love  Python    and   Regex"
```

```python
clean_text = re.sub(r"\s+"," ", text)
```

```python
clean_text
```
