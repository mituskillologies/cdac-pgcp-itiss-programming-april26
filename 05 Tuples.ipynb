```python
num = int(input("How many students? "))
stud = []
for i in range(num):
    roll = int(input("Enter roll: "))
    name = input("Enter name: ")
    marks = float(input("Enter marks: "))
    stud.append((roll, name, marks))

print("Your data:")
for i in range(num):
    print(stud[i])
```
