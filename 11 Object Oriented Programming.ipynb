#### Defining classes

```python
class Person: 
    def __init__(self, n, a):
        self.name = n
        self.age = a
    def show(self):
        print("Values are:")
        print(self.name, self.age)
```

```python
# create the object
p1 = Person("Ajay",34)
```

```python
p1.name
```

```python
p1.age
```

```python
p1.show()
```

```python
p2 = Person('Priya',25)
p2.show()
```

```python
class Person:
    city = 'Pune'           # class variable
    def __init__(self, n, a):  # constructor
        self.name = n       # instance variable
        self.age = a
    def show(self):         # instance method
        print("Values are:")
        print(self.name, self.age)
    def isEligible(self):   # instance method
        if self.age > 18: 
            return True
        else:
            return False
```

```python
p1 = Person('Ram', 16)
p2 = Person('Anita', 26)
```

```python
p1.isEligible()
```

```python
p2.isEligible()
```

```python
p1.city, p2.city
```

```python
class Person(object):        # base / super class
    def __init__(self, n, a=25):  # constructor
        self.name = n       # instance variable
        self.age = a
    def show(self):         # instance method
        print("Values are:")
        print(self.name, self.age)

class Employee(Person):     # derived / subclass
    def isEligible(self):   # instance method
        if self.age > 18: 
            return True
        else:
            return False
```

```python
e1 = Employee('Atul', 24)
```

```python
e1.show()
```

```python
e1.isEligible()
```

```python
e2 = Employee('Raja')
```

```python
e2.show()
```
