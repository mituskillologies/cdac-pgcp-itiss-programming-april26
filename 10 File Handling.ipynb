#### Example: File copy program

```python
# read data.txt and write in new_data.txt, show the contents
```

```python
f1 = open('data.txt')
f2 = open('new_data.txt', 'w')
f2.write(f1.read())
f2.close()

f2 = open('new_data.txt')
print(f2.read())
```
