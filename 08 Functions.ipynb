```python
def factorial(num): 
    fact = 1
    for i in range(1, num+1): 
        fact = fact * i 
    return fact
```

```python
factorial(5)
```

```python
n = 5
r = 3
factorial(n) / (factorial(r) * factorial(n-r))
```

#### Function returning multiple values

```python
def calc(n):
    sq = n ** 2
    cu = n ** 3
    return sq, cu
```

```python
calc(5)
```

```python
x, y = calc(5)
```

```python
x
```

```python
y
```

#### Default parameters and Keyword arguments

```python
def add(x=2,y=5,z=1): 
    addition = x + y + z
    return addition
```

```python
add(4,6,7)
```

```python
add(4,6)
```

```python
add(4)
```

```python
add()
```

```python
add(y=8)
```
