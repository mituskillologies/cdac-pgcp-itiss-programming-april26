#### While loop

```python
count  = 0
while count < 10: 
    print("Tushar")
    count += 1
```

```python
add = 0
num = 1
while num <= 10: 
    add += num 
    num += 1 
print(add)
```

#### break statement

```python
for i in range(10):
    if i == 5: 
       break
    print(i)
```

#### continue statement

```python
for i in range(10):
    if i == 5: 
       continue
    print(i)
```
